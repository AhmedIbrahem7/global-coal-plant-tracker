# Import Libraries

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
#get data from csv file
df=pd.read_csv('global-coal-plant-tracker.csv')
df
```

```python
#health check for data 
df.info()
```

```python
#check sumarry statistics of data
df.describe().round()
```

```python
#convert for new data type
df = df.convert_dtypes()
```

```python
#convert columns name to lower case
df.columns = df.columns.str.lower()
df.columns
```

```python
#check for dublicate data
df.duplicated().sum()
```

```python
#show duplicated data
df[df.duplicated()]
```

```python
#drop duplicated data
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
#check for missing values
df.isna().sum()
```

```python
#check missing vlaues %
df.isna().mean().round(4)*100
```

```python
# Data Cleaning
```

```python
#handle missing values
df.drop(columns=['sponsor'],inplace=True)
```

```python
#handle missing values in other columns
df = df.dropna(subset=['url','capacity_mw','status','region','country','geopoint'])
```

```python
df
```

```python
df.reset_index(inplace= True, drop= True)
```

```python
df.head()
```

```python
#check outliers
px.box(df, y='capacity_mw', title='Box Plot of Capacity (MW)')
```

```python
num_cols=df.select_dtypes(include= 'number').columns
num_cols
```

```python
#show data for each colmns with histogram and box plot
for col in num_cols:
    px.histogram(data_frame=df, x=col, title=col).show()
    px.box(data_frame=df, x=col, title=col).show()
```

```python
#1 What is the global distribution of coal plant statuses?
status_counts= df['status'].value_counts()
px.pie(values=status_counts.values, names=status_counts.index, title='Global Distribution of Coal')
```

```python
#2 Which countries have the highest coal power capacity?
country_capacity = df.groupby('country')['capacity_mw'].sum().sort_values(ascending=False).head(10)
px.bar(x=country_capacity.index, y=country_capacity.values,title='Top 10 Countries by Coal Power Capacity',text_auto=True, labels={'x':'Country', 'y':'Total Capacity (MW)',})
```

```python
#3 Which regions rely most on coal energy
region_capacity = df.groupby('region')['capacity_mw'].sum().sort_values(ascending=False)
px.bar(x=region_capacity.index, y=region_capacity.values,title='Coal Power Capacity by Region',text_auto=True, labels={'x':'Region', 'y':'Total Capacity (MW)',})
```

```python
#4 Which countries are retiring coal plants the fastest
retiring_plants = df[df['status'] == 'Retired']
retiring_capacity = retiring_plants.groupby('country')['capacity_mw'].sum().sort_values(ascending=False)
px.bar(x=retiring_capacity.index, y=retiring_capacity.values,title='Countries Retiring Coal Plants Fastest',text_auto=True, labels={'x':'Country', 'y':'Total Retiring Capacity (MW)',})
```

```python
#5 What is the average coal plant capacity by country
country_avg_capacity = df.groupby('country')['capacity_mw'].mean().sort_values(ascending=False)
px.bar(x=country_avg_capacity.index, y=country_avg_capacity.values,title='Average Coal Plant Capacity by Country',text_auto=True, labels={'x':'Country', 'y':'Average Capacity (MW)',})
```

```python
#6 Where are new coal plants under construction?
under_construction = df[df['status'] == 'Construction']
under_construction
```

```python
#How does operating vs retired capacity vary by region
capacity_by_region_status = df.groupby(['region', 'status'])['capacity_mw'].sum().reset_index()
capacity_by_region_status['percentage'] = (capacity_by_region_status.groupby('region')['capacity_mw'].transform(lambda x: x / x.sum()))
px.bar(capacity_by_region_status, x='region', y='percentage', color='status', title='Operating vs Retired Capacity by Region', text_auto='.2%')
```
